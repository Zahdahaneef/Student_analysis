# **3. Data Analysis** 

```python
# 1. Average final grade (G3)
average_G3 = data['G3'].mean()
print("Average final grade (G3) is:", average_G3)
```

```python
plt.figure(figsize=(8, 4))
plt.hist(data['G3'], bins=20, color='c', edgecolor='black')
plt.xlabel('Final Grade (G3)')
plt.ylabel('Frequency')
plt.title('Distribution of Final Grades (G3)')
plt.show()
```

* Most students scored between 10 and 15
* 
There are few very low scores (0–5) and few high scores (18–2
* Its a Slightly Normal Distribution0)tion

```python
# 2. How many students scored above 15?
count = (data['G3'] > 15).sum()
print("Number of students scoring above 15:", count)
```

```python
# 3. Is study time correlated with performance?
study_time_correlation1 = data['studytime'].corr(data['G1'])
print("Study time correlation with performance on G1:", study_time_correlation1)

study_time_correlation2 = data['studytime'].corr(data['G2'])
print("Study time correlation with performance on G2:", study_time_correlation2)

study_time_correlation3 = data['studytime'].corr(data['G3'])
print("Study time correlation with performance on G3:", study_time_correlation3)
```

* Here Result is 0 that means it has no corr relationship with performace and Studytime

```python
# 4. Which gender performs better on average?
avg_scores = data.groupby('sex')['G3'].mean()

print(avg_scores)
```

* Male students perform slightly better on average (10.91) compared to female students (9.97)

# **4. Visualizations**

**Histogram of grades**

```python
plt.figure()
plt.hist(data['G1'], bins=10,edgecolor='Black')
plt.title("G1 Distribution")
plt.xlabel("G1")
plt.ylabel("Frequency")
plt.show()

plt.figure()
plt.hist(data['G2'], bins=10,edgecolor='Black')
plt.title("G2 Distribution")
plt.xlabel("G2")
plt.ylabel("Frequency")
plt.show()

plt.figure()
plt.hist(data['G3'], bins=10, edgecolor='Black')
plt.title("G3 Distribution")
plt.xlabel("G3")
plt.ylabel("Frequency")
plt.show()
```

G1:
 * Most students have mid-range G1 scores (8–14), with few extremes.

G2:

* Most of the students have the highest scores between (8-15)

G3:
* Most of the students have the highest score between (10-12)
 


**Scatterplot: study time vs grades**

```python
plt.figure(figsize=(8,4))
plt.scatter(data['studytime'],data['G3'])
plt.title("Study Time vs Final Grade (G3)")
plt.xlabel("Study Time")
plt.ylabel("Final Grade")
plt.show()
```

* The scatter plot shows a slight positive relationship between study time and final grades. Students who study more tend to achieve higher scores, although the variation indicates other factors also influence performance.

```python
avg_scores.plot(kind='bar',figsize=(8,4))

plt.title("Average Performance by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Final Grade (G3)")

plt.show()
```
